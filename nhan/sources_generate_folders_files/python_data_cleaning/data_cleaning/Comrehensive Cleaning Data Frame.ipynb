```python
import email
import pandas as pd
data = {
    "customer_id":[1,2,3,4,4],
    "name":[" Alice ","Bob","Charlie","David","David"],
    "age":[25,None,35,40,40],
    "email":["alice@email.com","bob@email.com",None,"david@email.com","david@email.com"],
    "purchase_amount":["100","200","invalid","150","150"],
    "date":["2023-01-01","2023/02/01","March 3 2023","2023-04-01","2023-04-01"]
}

df = pd.DataFrame(data)
# Step 1 — Explore the Data
df.head()
df.info()
df.describe()
df.isnull().sum()
# Purpose:
# detect missing values
# check data types
# understand data distribution

# Step 2 — Remove Duplicates
df = df.drop_duplicates()

# Step 3 — Clean Text Columns
# Remove extra spaces.
df["name"] = df["name"].str.strip()

# Step 4 — Handle Missing Values
df["age"] = df["age"].fillna(df["age"].mean())
#Fill missing email with placeholder.
df["email"] = df["email"].fillna("unknown@email.com")

# Step 5 — Convert Data Types
# Convert purchase amount to numeric.
df["purchase_amount"] = pd.to_numeric(df["purchase_amount"], errors="coerce")
# Replace invalid values with median.
df["purchase_amount"] = df["purchase_amount"].fillna(df["purchase_amount"].median())

# Step 6 — Convert Date Format
# Use format='mixed' to handle different date formats
df["date"] = pd.to_datetime(df["date"], format='mixed')

# Step 7 — Detect Outliers
# Using IQR method.
Q1 = df["purchase_amount"].quantile(0.25)
Q3 = df["purchase_amount"].quantile(0.75)
IQR = Q3 - Q1

df = df[(df["purchase_amount"] >= Q1 - 1.5*IQR) & (df["purchase_amount"] <= Q3 + 1.5*IQR)]

# Step 8 — Create New Features
#  Example: purchase category.
df["purchase_category"] = df["purchase_amount"].apply(
    lambda x: "High" if x > 150 else "Low" )

#Step 9 — Sort Data
df = df.sort_values(by="purchase_amount", ascending=False)

# Step 10 — Save Clean Dataset
df.to_csv("clean_sales_data.csv", index=False)
```
